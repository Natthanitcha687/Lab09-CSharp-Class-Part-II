## คำถาม

1. Static Method คืออะไร


Static Method คือ เมธอดที่เรียกใช้ได้โดยตรงจากชื่อคลาส ไม่จำเป็นต้องสร้าง Object ของคลาสก่อน

2. ประกาศ static method ในภาษา C# อย่างไร


ใช้ keyword static นำหน้าชนิดข้อมูลที่ส่งกลับของเมธอด

3. เรียกใช้ static method อย่างไร


เรียกใช้โดยตรงจากชื่อคลาส เช่น


int result = MyClass.MyStaticMethod();


4. static method สามารถเข้าถึง instance members (fields, methods) ได้หรือไม่


ไม่ได้ Static Method สามารถเข้าถึงได้เฉพาะ Static Members เท่านั้น

5. static method เหมาะกับการใช้งานแบบไหน


-เมธอดที่ทำงานกับข้อมูลของคลาสโดยรวม

-เมธอด Utility ที่ไม่ผูกกับ Object ใดๆ

-เมธอดที่ทำงานกับค่าคงที่ของคลาส

6. ข้อดีของการใช้ static method คืออะไรบ้าง


-เรียกใช้ได้ง่ายและรวดเร็ว

-ไม่ต้องสร้าง Object ของคลาส

-ใช้หน่วยความจำอย่างมีประสิทธิภาพ

7. ข้อเสียของการใช้ static method คืออะไรบ้าง


-ไม่สามารถเข้าถึง Instance Members ได้

-อาจทำให้เกิดปัญหา Side Effect ถ้ามีการแก้ไขข้อมูล Static
