## คำถาม

1. ในภาษา C# นั้น Instance members และ static members  แตกต่างกันอย่างไร

Instance members (เช่น fields, methods) เป็นของแต่ละ object ที่สร้างจาก class

Static members (เช่น fields, methods) เป็นของ class โดยรวม ไม่ได้ผูกกับ object ใดๆ

2. เข้าถึง Instance members และ static members มีหลักการที่แตกต่างกันอย่างไร

Instance members เข้าถึงผ่าน object ที่สร้างจาก class (เช่น object.member)

Static members เข้าถึงโดยตรงผ่านชื่อ class (เช่น Class.member)

3. Instance members และ static members เหมาะกับการใช้งานแบบไหน

Instance members ใช้เก็บข้อมูลหรือพฤติกรรมที่แตกต่างกันไปในแต่ละ object

Static members ใช้เก็บข้อมูลหรือพฤติกรรมที่ใช้ร่วมกันในทุก object ของ class

4. Instance members และ static members ใช้หน่วยความจำต่างกันอย่างไร

Instance members ใช้หน่วยความจำเมื่อมีการสร้าง object

Static members ใช้หน่วยความจำครั้งเดียวเมื่อ class ถูกโหลด

5. Instance members และ static members มีอายุการใช้งานต่างกันอย่างไร

Instance members มีอายุการใช้งานตาม object ที่สร้างขึ้น

Static members มีอายุการใช้งานตั้งแต่ class ถูกโหลดจนกระทั่งโปรแกรมสิ้นสุด

6. Instance members และ static members สามารถใช้งานร่วมกันได้หรือไม่

Instance members และ static members สามารถใช้งานร่วมกันได้ใน class เดียวกันได้

7. keyword static ใช้กับสมาชิกประเภทใดของคลาสได้บ้าง

Keyword static ใช้กับ fields, methods และ constructors ของ class
