## คำถาม

1. เมื่อกำหนดให้สมาชิกในคลาสให้มี Access Modifier ดังต่อไปนี้ จะสามารถเข้าถึงจากที่ไหนได้บ้าง


1.1 public

ทุกที่ ทั้งภายในและภายนอกคลาส รวมถึงคลาสที่สืบทอดมา

1.2 private

เฉพาะภายในคลาสเดียวกันเท่านั้น

1.3 protected

ภายในคลาสเดียวกัน และคลาสที่สืบทอดมา (ถึงแม้จะอยู่คนละคลาส)

1.4 internal

ภายในโปรเจ็กต์ (Assembly) เดียวกันเท่านั้น

1.5 protected internal

ภายในโปรเจ็กต์เดียวกัน หรือคลาสที่สืบทอดมา (ถึงแม้จะอยู่นอกโปรเจ็กต์)
