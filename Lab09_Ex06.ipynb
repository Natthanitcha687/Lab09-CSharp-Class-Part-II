## คำถาม

1. Member constant คืออะไร และมีคุณสมบัติอย่างไร


เป็นค่าคงที่ที่ประกาศไว้ภายในคลาส มีค่าคงที่ตลอดอายุการใช้งานของโปรแกรม ไม่สามารถเปลี่ยนแปลงค่าได้หลังจากกำหนดค่าเริ่มต้นแล้ว

2. เราสามารถประกาศ member constant ในภาษา C# ได้อย่างไร


ใช้ keyword const ตามด้วยชนิดข้อมูล ชื่อ และค่า
เช่น public const int MAX_VALUE = 100;

3. member constant สามารถสร้างเป็นชนิดข้อมูลใดได้บ้าง


ใช้ได้กับชนิดข้อมูลพื้นฐานทุกชนิด เช่น int, float, string, bool
รวมถึง enum และ string

4. เราสามารถเข้าถึง member constant ได้อย่างไร


เข้าถึงผ่านชื่อคลาส (ถ้าเป็น static constant) หรือชื่อ object (ถ้าเป็น instance constant)
เช่น ClassName.MAX_VALUE หรือ objectName.MAX_VALUE

5. member constant เป็น static โดยปริยายหรือไม่


ไม่ member constant สามารถเป็นได้ทั้ง static (ใช้ร่วมกันในทุก object) หรือ instance (เฉพาะ object นั้นๆ)
ถ้าไม่ใส่ static จะเป็น instance constant

6. มีข้อจำกัดอะไรบ้างในการใช้ member constant 


ค่าต้องกำหนดตั้งแต่ตอนประกาศ และเปลี่ยนแปลงไม่ได้
ไม่สามารถใช้ static กับ instance constant และ vice versa

7. member constant เหมาะกับการใช้งานแบบไหน


เก็บค่าคงที่ที่ใช้ร่วมกันในโปรแกรม เช่น ค่าคงที่ทางคณิตศาสตร์, ข้อความ error, ขนาด array
กำหนดค่าเริ่มต้นที่ไม่ต้องการให้เปลี่ยนแปลง

8. จงตัวอย่าง Member Constant ในไลบรารี่ของภาษา C#

Math.PI (ค่าคงที่ Pi)

int.MaxValue (ค่าสูงสุดของชนิด int)
